```python
from transformers import pipeline
```

```python
qa_pipeline = pipeline(
    "question-answering",
    model="distilbert-base-cased-distilled-squad"
)
```

```python
context = """
Generative AI refers to artificial intelligence systems that can create new content such as text,
images, audio, and code. These models learn patterns from large datasets and use them to generate
realistic and meaningful outputs. However, risks include hallucinations, bias, and deepfakes.
"""
```

```python
question = "What is Generative AI?"

result = qa_pipeline(question=question, context=context)
print(result)
```

```python
context = input("Paste your study material:\n")

while True:
    question = input("\nAsk a question (type 'exit' to stop): ")
    if question.lower() == "exit":
        break

    answer = qa_pipeline(question=question, context=context)
    print("Answer:", answer["answer"])
    print("Confidence:", round(answer["score"], 3))
```

```python
from transformers import pipeline
```

```python
bert_model = "bert-base-uncased"
roberta_model = "roberta-base"
bart_model = "facebook/bart-base"
```

```python
prompt = "The future of Artificial Intelligence is"

# BERT
try:
    generator_bert = pipeline("text-generation", model=bert_model)
    print("BERT:", generator_bert(prompt, max_length=30))
except Exception as e:
    print("BERT Error:", e)

# RoBERTa
try:
    generator_roberta = pipeline("text-generation", model=roberta_model)
    print("RoBERTa:", generator_roberta(prompt, max_length=30))
except Exception as e:
    print("RoBERTa Error:", e)

# BART
generator_bart = pipeline("text-generation", model=bart_model)
print("BART:", generator_bart(prompt, max_length=30))
```

```python
sentence = "The goal of Generative AI is to [MASK] new content."

fill_mask_bert = pipeline("fill-mask", model=bert_model)
fill_mask_roberta = pipeline("fill-mask", model=roberta_model)
fill_mask_bart = pipeline("fill-mask", model=bart_model)

print("BERT:", fill_mask_bert(sentence))
print("RoBERTa:", fill_mask_roberta(sentence))
print("BART:", fill_mask_bart(sentence))
```

```python
context = "Generative AI poses significant risks such as hallucinations, bias, and deepfakes."
question = "What are the risks?"

qa_bert = pipeline("question-answering", model=bert_model)
qa_roberta = pipeline("question-answering", model=roberta_model)
qa_bart = pipeline("question-answering", model=bart_model)

print("BERT:", qa_bert(question=question, context=context))
print("RoBERTa:", qa_roberta(question=question, context=context))
print("BART:", qa_bart(question=question, context=context))
```
